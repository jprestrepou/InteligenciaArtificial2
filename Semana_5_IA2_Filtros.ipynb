```python
import matplotlib.pyplot as plt
import numpy as np

def convolucion(imagen, kernel):
    alto_imagen, ancho_imagen = len(imagen), len(imagen[0])
    alto_kernel, ancho_kernel = len(kernel), len(kernel[0])

    padding_alto = alto_kernel // 2
    padding_ancho = ancho_kernel // 2


    imagen_padding = [[0 for _ in range(ancho_imagen + 2 * padding_ancho)] for _ in range(alto_imagen + 2 * padding_alto)]

    for i in range(alto_imagen):
        for j in range(ancho_imagen):
            imagen_padding[i + padding_alto][j + padding_ancho] = imagen[i][j]

    resultado = [[0 for _ in range(ancho_imagen)] for _ in range(alto_imagen)]
    for i in range(alto_imagen):
        for j in range(ancho_imagen):
            suma = 0
            for m in range(alto_kernel):
                for n in range(ancho_kernel):
                    suma += imagen_padding[i + m][j + n] * kernel[m][n]
            resultado[i][j] = suma
    return resultado

imagen = [
    [1, 2, 3, 4, 5, 10],
    [4, 5, 6, 6, 7, 5],
    [7, 8, 9, 8, 9, 10],
    [7, 8, 9, 10, 11, 15],
    [7, 8, 9, 10, 12, 14],
    [7, 8, 9, 5, 10, 6],
]
kernel1 = [
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
]

kernel2 = [
    [-1, 0, -1],
    [-1, 0, -1],
    [-1, 0, -1]
]

kernel3 = [
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
]

kernel4 = [
    [0, 0, 0],
    [1, 1, 1],
    [0, 0, 0]
]

kernel5 = [
    [0, 0, 0],
    [-1, 1, -1],
    [0, 0, 0]
]

resultado_convolucion1 = convolucion(imagen, kernel1)
resultado_convolucion2 = convolucion(imagen, kernel2)
resultado_convolucion3 = convolucion(imagen, kernel3)
resultado_convolucion4 = convolucion(imagen, kernel4)
resultado_convolucion5 = convolucion(imagen, kernel5)

plt.subplot(2, 3, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Original')

plt.subplot(2, 3, 2)
plt.imshow(resultado_convolucion1, cmap='gray')
plt.title('Convolución')

plt.subplot(2, 3, 3)
plt.imshow(resultado_convolucion2, cmap='gray')
plt.title('Convolución')

plt.subplot(2, 3, 4)
plt.imshow(resultado_convolucion3, cmap='gray')
plt.title('Convolución')

plt.subplot(2, 3, 5)
plt.imshow(resultado_convolucion4, cmap='gray')
plt.title('Convolución')

plt.subplot(2, 3, 6)
plt.imshow(resultado_convolucion5, cmap='gray')
plt.title('Convolución')

plt.show()
```
